```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
import pandas_profiling as pp
import seaborn as sns
import warnings
import os
from pandas import DataFrame
from neo4j import GraphDatabase
from matplotlib.colors import ListedColormap
import squarify
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
from neo4j import GraphDatabase
class Neo4jConnection:
    
    def __init__(self, uri, user, pwd):
        self.__uri = uri
        self.__user = user
        self.__pwd = pwd
        self.__driver = None
        try:
            self.__driver = GraphDatabase.driver(self.__uri, auth=(self.__user, self.__pwd))
        except Exception as e:
            print("Failed to create the driver:", e)
        
    def close(self):
        if self.__driver is not None:
            self.__driver.close()
        
    def query(self, query, db=None):
        assert self.__driver is not None, "Driver not initialized!"
        session = None
        response = None
        try: 
            session = self.__driver.session(database=db) if db is not None else self.__driver.session() 
            response = list(session.run(query))
        except Exception as e:
            print("Query failed:", e)
        finally: 
            if session is not None:
                session.close()
        return response
```

```python
conn = Neo4jConnection(uri="bolt://localhost:7687", user="neo4j", pwd="root")
```

```python
query_string = '''
MATCH (player:Player)-[:HAS_APM]->(apm)
RETURN player.PlayerId, apm.IssueAPM
'''
dtf_data_1 = DataFrame([dict(_) for _ in conn.query(query_string, db='neo4j')])
dtf_data_1.columns = ['Player', 'IssueAPM']
```

### Player vs command (Most used command by player)

```python
q_comm = '''
MATCH (game:Game)-[:HAS_COMMAND]->(command:Command)
RETURN command.CommandName, count(game.GameId) AS count
ORDER BY count DESC
LIMIT 15
'''
df_comm = DataFrame([dict(_) for _ in conn.query(q_comm, db='neo4j')])
df_comm.columns = ['Command','Games']
```

```python
df_comm
```

```python
fig, ax = plt.subplots(figsize=(12, 9), subplot_kw=dict(aspect="equal"))

commands = df_comm.Command.to_list()


data = df_comm.Games.to_list()
explode = [0.05, 0.05, 0.05, 0.05, 0.05,0.05, 0.05, 0.05, 0.05,0.05,0.05, 0.05, 0.05, 0.05, 0.05]

wedges,texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-40,explode=explode)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(f"{commands[i]}\n{data[i]}", xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)
ax.legend(wedges, commands,
          title="Commands",
          loc="best",
          bbox_to_anchor=(1, 0, 0.5, 1))
ax.set_title("Top frequently used commands")

plt.show()
```

### Game vs start month and year

```python
q_start_year = '''
MATCH (game:Game)-[:HAS_START_YEAR]->(startyear:StartYear)
RETURN startyear.StartYear,count(game)
'''
df_start_year = DataFrame([dict(_) for _ in conn.query(q_start_year, db='neo4j')])
df_start_year.columns = ['StartYear','Games']
```
